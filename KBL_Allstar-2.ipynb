```python
#-*- coding:utf-8 -*-
import pandas as pd
import numpy as np
import sys
```

```python
# 데이터 불러오기
data = pd.read_csv("./17-18 basketball.csv")
team_code_data = pd.read_csv("./17-18 team_code.csv", encoding='euc-kr')
player_code_data = pd.read_csv("./17-18 player_code.csv", encoding='euc-kr')
```

```python
# 선수 코드 시리즈 생성

player_code = pd.Series(player_code_data['kname'])
player_code.index = (player_code_data['player_no'])
```

```python
# 선수 이름 매핑

name = []

for i in range(len(data)) :
    try:
        n = player_code[data.iloc[i]['player_no']]
        if type(n) == str :
            name.append(n)
        else :
            name.append(n.values[0])
    except:
        name.append(np.nan)
        
data['name'] = name
```

```python
# 팀 코드 시리즈 생성

team_code = pd.Series(team_code_data['team_name_1'])
team_code.index = team_code_data['team_code']
```

```python
# 팀 이름 매핑

team_name = []

for i in range(len(data)) :
    try:
        n = team_code[data.iloc[i]['team_code']]
        if type(n) == str :
            team_name.append(n)
        else :
            team_name.append(n.values[0])
    except:
        team_name.append(np.nan)
        
data['team_name'] = team_name
```

```python
# 결측치 제거

data = data.dropna(axis=0)
```

```python
len(data)
```

```python
# 외국인 제외

local_data = data[data['name'].apply(lambda x: len(x) <= 3)]
```

```python
len(local_data)
```

```python
column = data.columns[10:38]
playerData = []

for i in column :
    playerData.append(local_data.groupby(['team_name', 'name'])[i].sum())
    
df = pd.DataFrame(playerData)
df
```

```python
# 팀 토탈 시간

team_time_info = []
teamCode = local_data['team_name'].unique()
teamTime = []
team_in_players = []

for i in teamCode :
    teamTime.append((df[i][:].loc['play_min'].sum() * 60 + df[i][:].loc['play_sec'].sum()) / 60)
    team_in_players.append(len(df[i].columns))
    
team_info = pd.Series(teamTime, index=teamCode)
team_info_sort = team_info.sort_index().tolist()

for i,j in zip(team_info_sort, team_in_players) :
    for k in range(j) :
        team_time_info.append(i)
```

```python
# 팀 토탈 슛시도

team_fga_info = []
teamFga = []

for i in teamCode :
    teamFga.append((local_data.groupby(['team_name'])['fg_a'].sum())[i] + (local_data.groupby(['team_name'])['threep_a'].sum())[i])
    
team_info = pd.Series(teamFga, index=teamCode)
team_info_sort = team_info.sort_index().tolist()

for i,j in zip(team_info_sort, team_in_players) :
    for k in range(j) :
        team_fga_info.append(i)
```

```python
# 팀 토탈 자유투시도

team_fta_info = []
teamFta = []

for i in teamCode :
    teamFta.append((local_data.groupby(['team_name'])['ft_a'].sum())[i])
    
team_info = pd.Series(teamFta, index=teamCode)
team_info_sort = team_info.sort_index().tolist()

for i,j in zip(team_info_sort, team_in_players) :
    for k in range(j) :
        team_fta_info.append(i)
```

```python
# 팀 토탈 턴오버

team_to_info = []
teamTo = []

for i in teamCode :
    teamTo.append((local_data.groupby(['team_name'])['t_o'].sum())[i])
    
team_info = pd.Series(teamTo, index=teamCode)
team_info_sort = team_info.sort_index().tolist()

for i,j in zip(team_info_sort, team_in_players) :
    for k in range(j) :
        team_to_info.append(i)
```

```python
# 팀 토탈 어시스트

team_ast_info = []
teamAst = []

for i in teamCode :
    teamAst.append((local_data.groupby(['team_name'])['a_s'].sum())[i])
    
team_info = pd.Series(teamAst, index=teamCode)
team_info_sort = team_info.sort_index().tolist()

for i,j in zip(team_info_sort, team_in_players) :
    for k in range(j) :
        team_ast_info.append(i)
```

```python
# 팀 토탈 필드골

team_fg_info = []
teamFg = []

for i in teamCode :
    teamFg.append((local_data.groupby(['team_name'])['fg'].sum())[i])
    
team_info = pd.Series(teamFg, index=teamCode)
team_info_sort = team_info.sort_index().tolist()

for i,j in zip(team_info_sort, team_in_players) :
    for k in range(j) :
        team_fg_info.append(i)
```

```python
df.loc[len(df)] = team_time_info
df.loc[len(df)] = team_fga_info
df.loc[len(df)] = team_fta_info
df.loc[len(df)] = team_to_info
df.loc[len(df)] = team_ast_info
df.loc[len(df)] = team_fg_info
```

```python
df = df.rename(index={28 : 'team_time'})
df = df.rename(index={29 : 'team_fga'})
df = df.rename(index={30 : 'team_fta'})
df = df.rename(index={31 : 'team_to'})
df = df.rename(index={32 : 'team_ast'})
df = df.rename(index={33 : 'team_fg'})
```

```python
df = df.T.reset_index()
```

```python
df = df.set_index(df['name'])
```

```python
# usg 계산

usg = []


for i in range(len(df.index)) :
    field_goal_attempts = df.iloc[i]['fg_a']
    free_throw_attempts = df.iloc[i]['ft_a']
    turnover = df.iloc[i]['t_o']
    team_total_minute = df.iloc[i]['team_time']
    three_point_attempt = df.iloc[i]['threep_a']
    team_field_goal_attempt = df.iloc[i]['team_fga']
    team_free_throw_attempt = df.iloc[i]['team_fta']
    team_turnover_attempt = df.iloc[i]['team_to']
    play_time = (df.iloc[i]['play_min'] * 60 + df.iloc[i]['play_sec']) / 60
    
    usg.append(((field_goal_attempts + three_point_attempt
                 + 0.44 * free_throw_attempts + turnover)
                * team_total_minute) / ((team_field_goal_attempt +
                                         0.44 * team_free_throw_attempt +
                                         team_turnover_attempt)
                                        * 5 * play_time))
    
df['usg'] = usg
```

```python
# per parameter

league_fg = df['fg'].sum() #리그 필드골
league_ast = df['a_s'].sum() #리그 어시스트
league_ft = df['ft'].sum() #리그 자유투
league_score = df['score'].sum() #리그 득점
league_fga = df['fg_a'].sum() #리그 필드골 시도
league_fta = df['ft_a'].sum() #리그 자유투 시도
league_off_reb = df['o_r'].sum() #리그 공격리바운드
league_def_reb = df['d_r'].sum() #리그 수비리바운드
league_reb = league_off_reb + league_def_reb #리그 전체리바운드
league_to = df['t_o'].sum() #리그 턴오버
league_total_foul = df['foul_tot'].sum() #리그 파울
```

```python
# per 계산

per = []
factor = 2/3 - ((0.5*(league_ast/league_fg)) / (2*(league_fg/league_ft)))
vop = league_score / (league_fga-league_off_reb + league_to + 0.44*league_fta)
drbp = league_def_reb / league_reb


for i in range(len(df.index)) :
    three_point = df.iloc[i]['threep']
    assist = df.iloc[i]['a_s']
    team_assist = df.iloc[i]['team_ast']
    team_field_goal = df.iloc[i]['team_fg']
    turnover = df.iloc[i]['t_o']
    field_goal_attempts = df.iloc[i]['fg_a']
    field_goal = df.iloc[i]['fg']
    free_throw_attempts = df.iloc[i]['ft_a']
    free_throw = df.iloc[i]['ft']
    deffensive_rebound = df.iloc[i]['d_r']
    offensive_rebound = df.iloc[i]['o_r']
    steal = df.iloc[i]['s_t']
    block = df.iloc[i]['b_s']
    foul = df.iloc[i]['foul_tot']
    play_time = (df.iloc[i]['play_min'] * 60 + df.iloc[i]['play_sec']) / 60
    per.append(1/play_time*
               (three_point + 
                ((2/3)*assist) + ((2-factor*(team_assist/team_field_goal))*field_goal) 
                + (free_throw*0.5*(1+(1-(team_assist/team_field_goal))+(2/3)*(team_assist/team_field_goal)))
                - (vop*turnover) - (vop*drbp*(field_goal_attempts-field_goal))
                - (vop*0.44*(0.44+(0.56*drbp))*(free_throw_attempts-free_throw))
                + (vop*(1-drbp)*deffensive_rebound) + (vop*drbp*offensive_rebound) 
                + (vop*steal) + (vop*drbp*block) - (foul*((league_ft/league_total_foul)-0.44*(league_fta/league_total_foul)*vop))))
    
df['per'] = per
```
